## Load all evaluation scores

```python
with open(evaluation_result_file, 'r') as eval_f:
    lines = eval_f.readlines()
    for i in range(0, len(lines), metric_count):
        id = lines[i].split()[1]
        
        if id == 'all':
            continue
        
#         if int(id) > 868613:
#             continue
        
        dict = {}
        dict['q_id'] = id
        
        for j in range(metric_count):
            elements = lines[i + j].split()
            dict[elements[0]] = float(elements[2])
        evaluations.append(dict)

print("Finished loading evaluation")
```

## Sort evaluations on small to large

```python
sorted_map = sorted(evaluations, key=lambda elements: elements["map"])
filtered_queries = map(lambda x: [x["q_id"], x["P_100"]], evaluations)
filtered_queries = sorted(filtered_queries, key=lambda x: x[1])
# filtered_queries = list(filter(lambda x: x[1] > 0, filtered_queries))
filtered_queries = filtered_queries[:top_query_count]


print("Finished sorting, following queries are search for: ")
print(str(filtered_queries))
```
